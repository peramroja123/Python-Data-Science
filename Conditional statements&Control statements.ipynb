```python
#Perfect number
n=6
res=0
for i in range(1,n):
  if n%i==0:
    res=res+i
if n==res:
    print(f'{n} is a perfect number')
else:
    print(f'{n} is not a perfect number')
```

```python
#Break : we can stop the loop even if while condition is true
for i in range(11,21):
  print(i)
  if i==17:
    break
```

```python
#While loop: we can execute a set of statements as long as condition is true
a=1
while a<=10:
  print(a)
  a=a+1
```

```python
#Continue:used to end the current itteration in a loop and continues to next iteration
for i in range(11,17):
  if i==13:
    continue
  print(i)
```

```python
#Palindrome or not
n=121
b=str(n)
print(f'value in b:{b}')
c=b[::-1]
print(f'value in c:{c}')
if b==c:
  print('Palindrome number')
else:
  print('Not a palindrome')
```

```python
n=121
temp=n
r=0
while n>0:
  f=n%10
  r=(r*10)+f
  n=n//10
if temp==r:
  print('Palindrome')
else:
  print('Not a Palindrome')
```

```python
#Armstrong Number
n=153
temp=n
r=0
while n>0:
  f=n%10
  r=r+(f*f*f)
  n=n//10
if temp==r:
  print("Armstrong Number")
else:
  print("Not an Armstrong Number")
```

```python
#frequency of a character using dictionary concept
a='sharuk khan'
f={}
for i in a:
  if i not in f:
    f[i]=1
  else:
    f[i]=f[i]+1
print(f)
```

```python
#Add even and odd positions of a number->Microsoft
a=[7,9,8,9,0,1,0,9,9,6]
odd_based=a[::2]
even_based=a[1::2]
print(odd_based)
print(even_based)
c=0
d=0
for i in odd_based:
  c=c+i
for j in even_based:
  d=d+j
print(c)
print(d)
```

```python
a=[7,9,8,9,0,1,0,9,9,6]
odd_based=a[::2]
even_based=a[1::2]
print(odd_based)
print(even_based)
c=0
d=0
for i in range(5):
  c=c+odd_based[i]
  d=d+even_based[i]
print(c)
print(d)
```
