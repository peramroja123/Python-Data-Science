#Split and join functions in string

```python
#split
a='I love my country'
b=a.split()
print(b)
```

```python
a='i love my.country'
b=a.split('.')
print(b)
```

```python
#join
a=['usa','uk','paris','canada']
b=''.join(a)
print(b)
```

```python
b=' '.join(a)
print(b)
```

```python
a='i.love.programming.language'
b=a.split('.')
print(b)
c=(b[::-1])
print(c)
d='.'.join(c)
print(d)
```
