#append-it will take the single value from the user and will add to the existing list only at the end

```python
a=[10,50,3,13]
a.append(500)
print(a)
```

#Extend-It will take multiple values from the user and add to the excisting list only at the end

```python
a=[10,50,3,13]
a.extend([1,2])
print(a)
```

#Insert-we can add values whenever we want

```python
a=[10,50,3,13]
a.insert(2,300)
print(a)
```

#pop-pop will remove last value defaulty but if we pass a number it will treat like index

```python
 a=[10,50,3,13]
 a.pop()
 print(a)
```

```python
a=[10,50,3,13]
a.pop(2)
print(a)
```

#remove-will take value directly and remove the value

```python
a=[10,50,3,13]
a.remove(50)
print(a)
```

```python
#replace
a=[10,50,3,13]
a[1]=5
print(a)
```

#Count-Used to count the values

```python
a=[1,2,4,5,1,3,19,30,1]
print(a.count(1))
```

#index-used to find the index of a given value

```python
print(a.index(19))
```

```python
#del
a=[1,2,4,5]
del a
print(a)
```

```python
#clear
a=[1,2,4,5]
a.clear()
print(a)
```

```python
#copy---1.deep copy

#2.shallow copy
#Deep copy
a=[10,20,50]
b=a
print(a)
print(b)
a.append(100)
print(a)
print(b)
```

```python
#Shallow copy
a=[10,20,50]
b=a.copy()
print(a)
print(b)
a.append(100)
print(a)
print(b)
```

```python
#reverse
a=[10,20,30,40,50]
print(a[::-1])
```

```python
a=[10,20,30,40,50]
a.reverse()
print(a)
```

```python
#sort
a=[9,1,5,7,11,13,6,2,8]
a.sort()
print(a)
```

```python
a=[9,1,5,7,11,13,6,2,8]
a.sort(reverse=True)
print(a)
```
