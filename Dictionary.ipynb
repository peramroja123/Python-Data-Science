#Dict is Mutable

```python
#add
fruits={'apple':500,'carrot':100}
print(f'Before={fruits}')
fruits['apple']=1000
print(f'After={fruits}')
```

```python
fruits['banana']=1000
print(f'After={fruits}')
```

```python
fruits={}
print(fruits)
fruits['apple']=900
fruits['grapes']=200
print(fruits)
```

```python
#Delete
b={'apple': 500, 'carrot': 100, 'banana': 1000}
print(f'before:{b}')
del b['carrot']
print(f'After:{b}')
```

#Dict Built_in_functions
*get

*update

*pop

*popitem

*key

*value

*items

*clear

*copy


```python
#get-used to call the required data from the excisting dictionary
b={'apple': 1000, 'carrot': 100, 'banana': 1000}
print(b.get('carrot'))
```

```python
#Update:uesd to add key value to the excisting dictionary
b={'apple': 1000, 'carrot': 100, 'banana': 1000}
b.update({'grapes':200})
print(b)
```

```python
#pop:will take key from the user and will remove the key value
b={'apple': 1000, 'carrot': 100, 'banana': 1000}
b.pop('carrot')
print(b)
```

```python
#popitem : Will remove last key value
b={'apple': 1000, 'carrot': 100, 'banana': 1000}
b.popitem()
print(b)
```

```python
#Key: will remove keys
b={'apple': 1000, 'carrot': 100, 'banana': 1000}
print(b.keys())
```

```python
#Values : Will remove values
b={'apple': 1000, 'carrot': 100, 'banana': 1000}
print(b.values())
```

```python
#items
b={'apple': 1000, 'carrot': 100, 'banana': 1000}
print(b.items())
```

```python
#clear
b={'apple': 1000, 'carrot': 100, 'banana': 1000}
b.clear()
print(b)
```

```python
#copy
#deep copy
b={'apple': 1000, 'carrot': 100, 'banana': 1000}
a=b
print(a)
print(b)
print(id(a))
print(id(b))
```

```python
#Shallow copy
b={'apple': 1000, 'carrot': 100, 'banana': 1000}
a=b.copy()
print(a)
print(b)
print(id(a))
print(id(b))
```
