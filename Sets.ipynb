# Python sets are unordered pairs (Which means there will be no index concept in sets and no slicing&skipping also)

#Python sets wont allow duplicate values

```python
a={10,20,30,50,10,100,20,10}
print(a)
```

#Python sets are mutable but they dont allow mutable types inside it

```python
a={10,55.55,'data',(1,2,3)}
print(a)
```

#Python_built_in_functions
*Add
*Update
*pop
*discard
*remove
*clear
*copy
*union
*Intersection

```python
#add-will take only one input from the user and will to the existing set randomly
a={10,5,6,20,98,45}
print(a)
```

```python
print(f'Before adding:{a}')
a.add(600)
print(f'After adding:{a}')
```

```python
#update-will take multiple values from the user
a={10,5,6,20,98,45}
print(f'before:{a}')
a.update([100,200,300])
print(f'after:{a}')
```

```python
#pop-will remove one value randomly
a={10,5,6,20,98,45}
print(f'before{a}')
a.pop()
print(f'after:{a}')
```

```python
a.pop(2)
print(a)
```

```python
#Discard-will take only one input from the user and if the input is available it will remove or else it will be silent
a={10,5,6,20,98,45}
a.discard(98)
print(a)
```

```python
a={10,5,6,20,98,45}
a.discard(980)
print(a)
```

```python
#remove-will take only one input from the user and if the input is available it will remove or else it will through error
a={10,5,6,20,98,45}
print(f'before:{a}')
a.remove(98)
print(f'after:{a}')
```

```python
a={10,5,6,20,98,45}
print(f'before:{a}')
a.remove(980)
print(f'after:{a}')
```

```python
#Copy
#deep copy
a={10,5,6,20,98,45}
b=a
print(a)
print(b)
print(id(a))
print(id(b))
```

```python
a={10,5,6,20,98,45}
b=a.copy()
print(a)
print(b)
print(id(a))
print(id(b))
```

```python
#Union
a={1,2,3,4,5,6}
b={5,6,7,8,9,10}
print(a.union(b))
```

```python
#intersection
print(a.intersection(b))
```

```python
#clear
a={1,10,15,19,25}
a.clear()
print(a)
```
