# Before going to 2nd concept first we have to learn type conversions
Type conversions : Converting one data type into other data type
*int->float & int->str
*float->int & float->str
*str->int & str->float

```python
#int->float
a=21
print(a)
print(type(a))

b=float(a)#actual conversion is going on
print(b)
print(type(b))
```

```python
#int->str
a=21
print(a)
print(type(a))

b=str(a)#actual conversion is going on
print(b)
print(type(b))
```

```python
#float->int

a=21.2654
print(a)
print(type(a))

b=int(a)#actual conversion is going on
print(b)
print(type(b))
```

```python
#float->str
a=21.2654
print(a)
print(type(a))

b=str(a)#actual conversion is going on
print(b)
print(type(b))
```

```python
#str->int
a='21'
print(a)
print(type(a))

b=int(a)#actual conversion is going on
print(b)
print(type(b))
```

```python
#str->int
a='python123'
print(a)
print(type(a))

b=int(a)#actual conversion is going on
print(b)
print(type(b))
```

```python
#str->float
a='21.98'
print(a)
print(type(a))

b=float(a)#actual conversion is going on
print(b)
print(type(b))
```

2.)Using + we can add string data type along with given variables but make sure that variables are also string and + operator wont give tab space

```python
a=21
print('my age is '+str(a))
```

```python
a=45
b='USA'
c=90.022
#I am at the age of 45 staying in usa and earning 90.22 thousand dollars per year.
print('I am at the age of '+str(a)+' staying in '+str(b)+' and earning '+str(c)+' thousand dollars per year')
```

3.) Using Identifiers -> %d->int
                         &%f->float
                         &%s->str

```python
a=21
print('my age is %d'%(a))
```

```python
a=45
b='USA'
c=90.022
#I am at the age of 45 staying in usa and earning 90.22 thousand dollars per year.
print('I am at the age of %d staying in %s and earning %f thousand dollars per year'%(a,b,c))
```

4.) using {} and f function

```python
a=21
print(f'my age is {a}')
```

```python
a=45
b='USA'
c=90.022
#I am at the age of 45 staying in usa and earning 90.22 thousand dollars per year.
print(f'I am at the age of {a} staying in {b} and earning {c} thousand dollars per year')
```
